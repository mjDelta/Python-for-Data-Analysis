### random walk

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline

position=0
walk=[position]
steps=1000
for s in range(steps):
    step=1 if random.randint(0,1) else -1
    position+=step
    walk.append(position)

plt.plot(range(steps+1),walk)
plt.show()
```

### batch random walk

```python
import numpy as np
nwalks=500
steps=1000
draws=np.random.randint(0,2,size=(nwalks,steps))

draws=np.where(draws==0,-1,1)
walks=draws.cumsum(1)
```

```python
hit30s=(np.abs(walks)>=30).any(1)
hit30s_walks=walks[hit30s]

cross_times=(np.abs(hit30s_walks)>30).argmax(1)
cross_times
```

```python
cross_times.mean()
```
